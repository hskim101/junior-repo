## 2) 입력 받은 지역의 일일 증감률 시각화

```python
df
```

```python
# input area to draw plot
area=input()
result=df[df['지역']==area]
result
```

```python
# 1번과 동일하게 일일확진자 수 열 추가
result['확진자수2']=0
result['확진자수2'][1:]=result['확진자수'][:-1]

result['일일확진자수']=result['확진자수']-result['확진자수2']
result
```

```python
# 증감률 열 추가 후 0으로 초기화
result['증감률']=0
result.loc[result['날짜']=='2020.03.02','일일확진자수']=0
```

```python
result
```

```python
# 증감률 계산 (오늘 일일확진자수 / 전날까지 누적확진자 수)*100
for i in range(1,len(result)):
    result.iloc[i,5]=result.iloc[i,4]/result.iloc[i,3]*100
result
```

```python
# plot 그리기 1번과 동일
plt.rc('font',family='Malgun Gothic')
plt.title(area)
plt.plot(result['날짜'],result['증감률'])
plt.xlabel("날짜")
plt.ylabel("증감률")
plt.xticks(['2020.03.02','2020.04.21','2020.06.10','2020.07.30','2020.09.18','2020.11.07'])
```

## 3) 지역 일일 평균 확진자 수가 가장 많은 지역 3곳 시각화(입력 지역 x)

```python
df
```

```python
# data 원본 보호를 위해 새로운 변수에 deep copy 
result=df.copy()
```

```python
result
```

```python
result['일일확진자수']=0
result
```

```python
result=result.sort_values(by=['지역','날짜'])
```

```python
result['확진자수2']=0
result['확진자수2'][1:]=result['확진자수'][:-1]
result
```

```python
result.loc[result['날짜']=='2020.03.02','확진자수2']=0
result
```

```python
result['일일확진자수']=result['확진자수']-result['확진자수2']
result
```

```python
# 각 지역의 첫 날 일일확진자수 0으로 설정
result.loc[result['날짜']=='2020.03.02','일일확진자수']=0
result
```

```python
# 일일 평균 확진자수가 많은 순서대로 Top 3를 고르기 위해 딕셔너리에 저장
means={}
for area in result['지역'].unique():
    means[area]=result[result['지역']==area]['일일확진자수'].mean()
means
```

```python
means_df=pd.DataFrame.from_dict(means, orient='index').rename(columns={0:'평균'})
means_df=means_df.sort_values(by='평균',ascending=False)
means_df
```

```python
# Top 3 지역 선택
answer=[]
for i in means_df.index[:3]:
    answer.append(i)
answer
```

```python
plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('Top-3')

for i in answer:
    j=result[result['지역']==i].sort_values(by='날짜') # 시간 순서대로 plot을 그리기 위해 다시 날짜 오름차순으로 정렬
    label=i # legend를 표시하기 위해 list 안에 문자열로 저장되어 있는 값을 추출 
    plt.plot(j['날짜'],j['일일확진자수'], label=label)
plt.xlabel("날짜")
plt.xticks(['2020.03.02','2020.04.21','2020.06.10','2020.07.30','2020.09.18','2020.11.07'])
plt.legend()
```
