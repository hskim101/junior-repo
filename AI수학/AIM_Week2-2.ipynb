```python
import numpy as np
```

## Inverse matrix ($\mathbf{A^{-1}}$)

```python
A = np.array([[3, 2, 0],
              [-1, -3, 6],
              [2, 3, -5]])

print(A)
```

```python
invA = np.linalg.inv(A)
print(invA)
```

## Linear System

```python
X = np.array([[3, 1, 2], 
              [2, 6, -1], 
              [4, 0, -1]])
y = np.array([[5], 
              [1], 
              [3]])
print(X)
print(y)
```

```python
sol = np.linalg.solve(X, y)

print(sol)
```
